# **(b)** #
Two Preprocessing Steps:

Label Encoding: Convert the categorical target Final_Result (Pass/Fail) into numerical values (e.g., 1 and 0). This is necessary because most machine learning algorithms require numerical inputs.
Feature Scaling (Normalization/Standardization): The features are all percentages or averages but might have different spreads. Scaling them to a range (like 0-1) ensures that features with larger numerical ranges don't dominate the model training.
