AI (specifically supervised machine-learning and time-series forecasting models) is ideal because it can:

Learn hidden patterns from historical distribution data, past Hari Raya records, family feedback, and external variables.
Provide daily demand forecasts with quantified uncertainty (e.g., “85 % confidence that 220–240 lunchboxes will be needed tomorrow”).
Continuously improve accuracy as new data arrives (online learning).
Optimise resource allocation automatically, reducing waste by 20–40 % (common benchmark in AI-driven demand forecasting).
Scale effortlessly to 40 families × 30 days while incorporating real-time inputs such as weather APIs or festival event calendars.
In short, AI turns an uncertain, high-stakes logistics problem into a data-driven, low-waste operation that maximises the charity’s limited budget and impact.


```python
import random
import datetime

# =============================================
# Charity Organisation Food Demand Prediction
# CompanyA - Hari Raya Lunchbox Distribution
# =============================================

# Parameters
num_families = 40
num_days = 30
avg_family_size = 4.0  # Average Malaysian household size ~3.8-4.0 persons

# Food options with realistic prices (MYR per lunchbox, 2025-2026 estimates)
food_options = {
    "Nasi Lemak": {"price": 6.50, "description": "Classic coconut rice with sambal, egg, anchovies"},
    "Nasi Ayam Rendang": {"price": 12.50, "description": "Fragrant rice with tender chicken rendang"},
    "Halal Mixed": {"price": 9.50, "description": "Assorted halal dishes (rice + protein + vegetables)"}
}

# Generate dates starting from today (or around Hari Raya period)
start_date = datetime.date.today()
dates = [start_date + datetime.timedelta(days=i) for i in range(num_days)]

# Simulate daily demand with some variation (AI prediction model)
# Base: 40 families * ~1.1 lunchboxes per family (some families may need extra for guests/kids)
base_daily_lunchboxes = int(num_families * 1.1)

print("=" * 60)
print("CompanyA AI Food Demand Prediction - Hari Raya Lunchbox Program")
print(f"Beneficiaries: {num_families} families")
print(f"Duration: {num_days} days")
print(f"Estimated Average Family Size: {avg_family_size} persons")
print(f"Total Estimated People Served: {num_families * avg_family_size:.0f}")
print("=" * 60)
print()

total_cost = 0
daily_records = []

for day, date in enumerate(dates, 1):
    # Simulate daily variation: +/- 15% due to festive mood, weather, or attendance
    variation = random.uniform(0.85, 1.15)
    daily_lunchboxes = max(30, int(base_daily_lunchboxes * variation))

    # Randomly choose food type for the day (or mix)
    food_type = random.choice(list(food_options.keys()))
    price_per_box = food_options[food_type]["price"]

    daily_cost = daily_lunchboxes * price_per_box

    # Record data
    daily_records.append({
        "day": day,
        "date": date.strftime("%Y-%m-%d"),
        "weekday": date.strftime("%A"),
        "food_type": food_type,
        "lunchboxes": daily_lunchboxes,
        "price_per_box": price_per_box,
        "daily_cost": daily_cost
    })

    total_cost += daily_cost

    print(f"Day {day:2d} ({date.strftime('%Y-%m-%d')} - {date.strftime('%A')})")
    print(f"  Food: {food_type} ({food_options[food_type]['description']})")
    print(f"  Lunchboxes distributed: {daily_lunchboxes}")
    print(f"  Cost today: RM {daily_cost:.2f} (RM {price_per_box:.2f} each)")
    print("-" * 50)

# Summary
print("\n" + "=" * 60)
print("SUMMARY REPORT")
print("=" * 60)
print(f"Total lunchboxes distributed over 30 days : {sum(rec['lunchboxes'] for rec in daily_records)}")
print(f"Total estimated cost                    : RM {total_cost:.2f}")
print(f"Average lunchboxes per day              : {sum(rec['lunchboxes'] for rec in daily_records)/num_days:.1f}")
print(f"Average cost per day                    : RM {total_cost/num_days:.2f}")
print(f"Average family size used in planning    : {avg_family_size} persons")
print(f"Total people approximately served       : {num_families * avg_family_size * 30 / 30:.0f} (cumulative reach)")

# Optional: Save to CSV (uncomment if needed)
# import pandas as pd
# df = pd.DataFrame(daily_records)
# df.to_csv("hari_raya_lunchbox_prediction.csv", index=False)
# print("\nDetailed report saved to 'hari_raya_lunchbox_prediction.csv'")

print("\nScript completed. This simulation can be enhanced with real AI models (e.g., time-series forecasting with Prophet or LSTM) using historical demand data.")
```
